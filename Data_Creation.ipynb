```python
def wait(driver,xpath):
        element = WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.XPATH, xpath)))

#17 Kasım 2022 Perşembe, 08:40
def dateConversion(inp):
        processedInp = inp.replace(',', '')
        splitted = processedInp.split(' ')
        day = int(splitted[0])
        month_str = splitted[1]
        year = int(splitted[2])
        
        month  = -1
        if(month_str == 'Ocak'):
            month = 1
        elif(month_str == 'Şubat'):
            month = 2
        elif(month_str == 'Mart'):
            month = 3
        elif(month_str == 'Nisan'):
            month = 4
        elif(month_str == 'Mayıs'):
            month = 5
        elif(month_str == 'Haziran'):
            month = 6
        elif(month_str == 'Temmuz'):
            month = 7
        elif(month_str == 'Ağustos'):
            month = 8
        elif(month_str == 'Eylül'):
            month = 9
        elif(month_str == 'Ekim'):
            month = 10
        elif(month_str == 'Kasım'):
            month = 11
        elif(month_str == 'Aralık'):
            month = 12
            
        return datetime.datetime(year, month, day)
    

driver_path = "C:\\Users\\suley\\Desktop\\YAP470\\Proje\\Proje\\chromedriver.exe"
page = 'https://www.bloomberght.com/tum-ekonomi-haberleri'


titles_xpath = "//li//figcaption//*[@class = 'title']"
des_xpath = "//li//figcaption//*[@class = 'description']"
dates_xpath = "//li//figcaption//*[@class = 'date']"
sonraki_btn = "//span[text() = 'Sonraki']"

chrome_options = Options()
chrome_options.add_argument('--disable-notifications')

driver = webdriver.Chrome(executable_path=driver_path,chrome_options=chrome_options)
driver.get(page)

news_df = pd.DataFrame(columns=['Title', 'Description', 'Date', 'Source'])
work = True
from_date = datetime.datetime(2022, 8, 1)
to_date = datetime.datetime(2022, 10, 31)

start = timeit.default_timer()
while(work):
    wait(driver,titles_xpath)
    wait(driver,des_xpath)
    wait(driver,dates_xpath)
    titles = driver.find_elements(By.XPATH, titles_xpath)
    desc = driver.find_elements(By.XPATH, des_xpath)
    dates = driver.find_elements(By.XPATH, dates_xpath)
    for i in range(len(titles)):
        
        date = dateConversion(dates[i].text)

        if(date>to_date):
            continue
        if(date < from_date):
            work = False
            break
        print(f'Current Date:{date}, Title: {titles[i].text}')    
        news_df.loc[len(news_df.index)] = [titles[i].text, desc[i].text, date, 'BloombergHT']  
        
    wait(driver,sonraki_btn)
    sonraki = driver.find_element(By.XPATH, sonraki_btn)
    driver.execute_script("arguments[0].click();", sonraki)
stop = timeit.default_timer()    
driver.quit()
print(f'Done succesfully in {(stop - start)/60} minutes')


        
```
